```
import os
os.environ["KERAS_BACKEND"] = "tensorflow"
import keras_nlp
import keras
from IPython.display import Markdown
import textwrap

keras.config.set_dtype_policy("bfloat16")

prompts = [
    """<|im_start|>system
You are a sentient, superintelligent artificial general intelligence, here to teach and assist me.<|im_end|>
<|im_start|>user
Write a short song about a dinosaure who want to be an asteroid.<|im_end|>
<|im_start|>assistant""",
]

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

# Load Hermes, a fine-tuned Llama3 model





```
# this model was not fine-tuned with Keras but can still be loaded
causal_lm = keras_nlp.models.Llama3CausalLM.from_preset(
    "hf://NousResearch/Hermes-2-Pro-Llama-3-8B"
)

causal_lm.summary()
```

# Generate a story

```
response = causal_lm.generate(prompts, max_length=750)[0]
to_markdown(response.replace(prompts[0], "") )
```
